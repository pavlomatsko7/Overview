```python
import pandas as pd
dff = pd.read_csv(r'C:\Users\Pavlo\Desktop\EEM.csv', parse_dates=['Date'], index_col='Date')
df = pd.DataFrame(dff['Close'])
df.head()
```

```python
df['Prev Day Price'] = df['Close'].shift(1)
df
```

```python
#Calculating the diff from current day to prev day
df['1 day change'] = df['Close'] - df['Prev Day Price']
df
```

```python
#Calculating 5 day % return
df['5 day % return'] = (df['Close'] - df['Close'].shift(5))*100/df['Close'].shift(5)
df
```

```python
# just showing the Close price and removes all other columns
df = df[['Close']]
df
```
